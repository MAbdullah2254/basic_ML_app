```python
# Importing necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# Reading Algerian_forest_fires_dataset_UPDATE.csv file
orignal_dataset=pd.read_csv("/content/sample_data/Algerian_forest_fires_dataset_UPDATE.csv", header=1)
```

```python
orignal_dataset.head()
```

```python
# Getting necessary information about the dataset
orignal_dataset.info()
# After checking it turned out that most of the features gor 'object' datatype that is needed to convert into int or float datatype for model training.
```

**Data Cleaning**

```python
# Finding any missing values
orignal_dataset[orignal_dataset.isnull().any(axis=1)]
```

The dataset is converted into two sets based on region from 122th index,we

make a new column based on region

1 : "Bajaia Region Dataset"

2 : "Sidi-Bel Abbes Region Dataset"

Add new column with region



```python
orignal_dataset.loc[:122,"Region"]=0
orignal_dataset.loc[122:,"Region"]=1
```

```python
orignal_dataset.info()
```

```python
# To avoid saving memory for float, converting "Region" column to int
orignal_dataset["Region"]=orignal_dataset["Region"].astype(int)
```

```python
orignal_dataset.head()
```

```python
# Removing the null values
orignal_dataset=orignal_dataset.dropna().reset_index(drop=True)
```

```python
orignal_dataset.isnull().sum()
```

```python
orignal_dataset.iloc[[122]]
```

```python
# Drop 122nd row
orignal_dataset=orignal_dataset.drop(122).reset_index(drop=True)
```

```python
orignal_dataset.iloc[[122]]
```

```python
# Fix spaces in column names
orignal_dataset.columns = orignal_dataset.columns.str.strip()
orignal_dataset.columns
```

**Changing Datatypes of required columns**

```python
# Convert columns day-Ws into int
orignal_dataset[["day","month","year","Temperature","RH","Ws"]]=orignal_dataset[["day","month","year","Temperature","RH","Ws"]].astype(int)
orignal_dataset.info()
```

```python
# Changing the other columns' datatype into float except "Classes and Region"
objects=[features for features in orignal_dataset.columns if orignal_dataset[features].dtype=='O']
for object in objects:
  if object!="Classes":
    orignal_dataset[object]=orignal_dataset[object].astype(float)
```

```python
orignal_dataset.info()
```

```python
# Save the cleaned dataset into a new file
orignal_dataset.to_csv("Algerian_forest_fires_cleaned_dataset.csv",index=True)
```

```python
# Region wise monthly fire analysis
df_copy1=orignal_dataset
df_copy1["Classes"] = df_copy1["Classes"].astype(str).str.strip().str.lower()
df_copy1["Classes"]=np.where(df_copy1["Classes"]=="fire","fire","not fire")
df_copy1.head()
```

```python
dfregion1=df_copy1.loc[df_copy1["Region"]==1]
plt.subplots(figsize=(13,6))
sns.set_style('whitegrid')
sns.countplot(x="month",hue="Classes",data=dfregion1)
plt.title("Sidi-Bel Albes Region Monthly Fire Analysis", weight="bold")
plt.ylabel("Number of Fires",weight="bold")
plt.xlabel("Month",weight="bold")
plt.legend(title="Classes",loc="upper right")
plt.show()
```

```python
dfregion2=df_copy1.loc[df_copy1["Region"]==0]
plt.subplots(figsize=(13,6))
sns.set_style('whitegrid')
sns.countplot(x="month",hue="Classes",data=dfregion2)
plt.title("Bajaia Region Monthly Fire Analysis", weight="bold")
plt.ylabel("Number of Fires",weight="bold")
plt.xlabel("Month",weight="bold")
plt.legend(title="Classes",loc="upper right")
plt.show()
```

From the monthly fire analysis of both the regions it is concluded that they had huge number of fires in August.

Overall number of fires is huge in the months of June,July and August.

Semptember happened to be the month of less fires for both the regions.

Sidi-Bel Albes region had huge fires in Semptember as compare to Bajaia region that has significantly less fires in the same month.
